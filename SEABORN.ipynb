1 **histplot()**
* histogram plot

* kde(kernal distribution density) : is a technique to estimate the peak probablity distribution of a data set smoothly

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a=pd.read_csv("/content/sales_dataset.csv")
a.head()
```

```python
sns.histplot(data=a,x='TotalSales',bins=20,kde=True,color='turquoise')
#kde(kernal distribution density) : is a technique to estimate the peak probablity distribution of a data set smoothly
plt.title("Distribution of Total Sales")
plt.show()
```

# Note : outliers ---> The data which is present outside the box plot boundary

**IQR-inter quaratile range**

* Quartile: The contribution of a value in visualization
* They are two quartile values which is Q3 having 75% contribution to the visualization where Q1 cpntibutes 25%
* Note : IQR works only when the given collection of values are sorted(ascending)
* Formula to calculate IQR values: Q3-Q1

```python
q1=b['alcohol'].quantile(0.25)
q3=b['alcohol'].quantile(0.75)

iqr=q3-q1
print(q1,',',q3,',',iqr)
```

**To calculate lower bound and upper bound**

```python
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

print(lower_bound)
print(upper_bound)
```

**To calculate outliers**

```python
outliers = b[(b['alcohol'] < lower_bound) | (b['alcohol'] > upper_bound)]
outliers
```

# **Winsorization /Capping**
* it is used to replace the values beyond bounds with nearest values


```python
winzorised_data=b['alcohol'].clip(lower=lower_bound,upper=upper_bound)
print(winzorised_data)
```

# **Regression Plot**
* it is also called scatter plot with a regression line
* Note : it is used to visualize the relationship between two numerical variables and optionally two fit a regression model
* the regression line shows the best fit line through the data often computed using linear regression
* Syntax: sns.regplot(x='col1',y='col2',data=var)
    * x- cloumn name for x axis
    * y- cloumn name for y axis

```python
sns.regplot(x='alcohol',y='quality',data=b)
plt.show()
```

scatter_kws
* it is used to maintain the size and transperency of points

```python
sns.regplot(x='alcohol',y='quality',data=b,scatter_kws={'s':100,'alpha':0.5})
plt.show()
```

line_kws
* it is used to maintain the size and transperency of lines

```python
sns.regplot(x='alcohol',y='quality',data=b,scatter_kws={'s':100,'alpha':0.5},line_kws={'color':'red','linewidth':0.5})
plt.show()
```
